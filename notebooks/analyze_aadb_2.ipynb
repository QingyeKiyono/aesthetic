```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
aadb_data = pd.read_csv("../data/aadb.csv")
aadb_data.head(1)
```

```python
labels = ["dof", "balancing_elements", "color_harmony", "content", "light", "motion_blur", "object", "repetition",
          "symmetry", "rule_of_thirds", "vivid_color"]

a = list((aadb_data[labels[i]] == 1).sum() for i in range(11))
a
```

```python
ax = sns.barplot(x=labels, y=a)
ax.set_xticklabels(labels=labels, rotation=45, fontsize=15)
ax.figure.set_size_inches(10, 5)
plt.xlabel("label name", fontsize=15)
plt.ylabel("count", fontsize=15)
```
